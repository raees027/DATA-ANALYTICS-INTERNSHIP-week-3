Z-Score method for finding outlieres

```python
upper_limit=df6['Age'].mean()+3*df6['Age'].std()
lower_limit=df6['Age'].mean()-3*df6['Age'].std()
upper_limit,lower_limit
```

```python
df6.loc[(df6['Age']>upper_limit) | (df6['Age']<lower_limit)]
```

```python
sns.boxplot(df6['Age'])
```

```python
upper_limit=df6['Rating'].mean()+3*df6['Rating'].std()
lower_limit=df6['Rating'].mean()-3*df6['Rating'].std()
upper_limit,lower_limit
```

```python
df6.loc[(df6['Rating']>upper_limit) | (df6['Rating']<lower_limit)]
```

```python
q1=df6['Rating'].quantile(0.25)
q3=df6['Rating'].quantile(0.75)
iqr=q3-q1
```

```python
q1,q3,iqr
```

```python
upper_limit=q3+(1.5*iqr)
lower_limit=q3-(1.5*iqr)
upper_limit,lower_limit
```

```python
sns.boxplot(df6['Rating'])
```
